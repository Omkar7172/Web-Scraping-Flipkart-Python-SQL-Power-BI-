```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []

for i in range(1, 3):  # Only first page for demo
    url = f"https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&param=19873&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIk1vdG9yb2xhIHNtYXJ0cGhvbmVzIl0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&wid=20.productCard.PMU_V2_12&sort=popularity&page={i}"
    
    headers = {
         'User-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, "html.parser")
        
        products = soup.find_all("div", class_="cPHDOP col-12-12")

        for product in products:
            # Basic Info
            title = product.find("div", class_="KzDlHZ")
            price = product.find("div", class_="Nx9bqj _4b5DiR")
            rating = product.find("div", class_="_5OesEi")
            
            # Specifications
            specs = product.find("ul", class_="G4BRas")
            spec_list = [li.text.strip() for li in specs.find_all("li")] if specs else []
            
            # Create dictionary with all fields (including empty ones)
            product_data = {
                "Product": title.text.strip() if title else "NA",
                "Price": price.text.strip() if price else "NA",
                "Rating": rating.text.strip() if rating else "NA",
                "Storage": spec_list[0] if len(spec_list) > 0 else "NA",
                "Display": spec_list[1] if len(spec_list) > 1 else "NA",
                "Camera": spec_list[2] if len(spec_list) > 2 else "NA",
                "Battery": spec_list[3] if len(spec_list) > 3 else "NA",
                "Processor": spec_list[4] if len(spec_list) > 4 else "NA",
                "Warranty": spec_list[5] if len(spec_list) > 5 else "NA"
            }
            
            data.append(product_data)
            print("Scraped:", product_data["Product"])

    except Exception as e:
        print(f"Error on page {i}: {str(e)}")

# Create and save DataFrame
if data:
    product_df = pd.DataFrame(data)
    product_df.to_csv("flipkart_products_detailed.csv", index=False)
    print("\nSaved to flipkart_products_detailed.csv")
    print(product_df.head())
else:
    print("No data scraped. Check selectors or website structure.")
```
